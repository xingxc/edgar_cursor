# SEC Edgar Package

## 1. Return the CIK and ticker relationship


### 1.1 Import packages

```python
import requests
import pandas as pd
```

### 1.2 Return CIK and Ticker relationship.
- Response [200]: means successfull response.


```python
header = {"User-Agent": "email@address.com"}

company_tickers = requests.get(
    "https://www.sec.gov/files/company_tickers.json", headers=header
)

df_tickers = pd.DataFrame.from_dict(company_tickers.json(), orient="index")
df_tickers["cik_str"] = df_tickers["cik_str"].astype(str).str.zfill(10)

display(company_tickers)
print(f"returned to tickers to __df_tickers")
display(df_tickers)
```

## Get the company facts

### 1. First return the CIK by searching through DataFrame for the relative ticker

```python
ticker = 'GS'
cik = df_tickers[df_tickers["ticker"] == ticker]["cik_str"].iloc[0]
display(f"Ticker-{ticker} ; CIK-{cik}")
```

### 2. Return the company facts

- Response [200]: means successfull response.


```python
company_facts = requests.get(
    f"https://data.sec.gov/api/xbrl/companyfacts/CIK{cik}.json",
    headers=header,
)

display(company_facts)
```
